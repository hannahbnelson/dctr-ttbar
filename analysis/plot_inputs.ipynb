```python
smeft_histo_dict = {
    "avg_top_pt" : Hist(hist.axis.Regular(bins=20, start=0, stop=400, name="avg_top_pt", label='avg_top_pt')),
    "mtt"        : Hist(hist.axis.Regular(bins=75, start=0, stop=1500, name="mtt", label='mtt')),
    "top1pt"     : Hist(hist.axis.Regular(bins=40, start=0, stop=800, name="top1pt", label='top1pt')),
    "top1eta"    : Hist(hist.axis.Regular(bins=40, start=-4, stop=4, name='top1eta', label='top1eta')),
    "top1phi"    : Hist(hist.axis.Regular(bins=40, start=-4, stop=4, name='top1phi', label="top1phi")),
    "top1mass"   : Hist(hist.axis.Regular(bins=20, start=50, stop=250, name='top1mass', label="top1mass")),
    "top2pt"     : Hist(hist.axis.Regular(bins=40, start=0, stop=800, name="top2pt", label='top2pt')),
    "top2eta"    : Hist(hist.axis.Regular(bins=40, start=-4, stop=4, name='top2eta', label='top2eta')),
    "top2phi"    : Hist(hist.axis.Regular(bins=40, start=-4, stop=4, name='top2phi', label="top2phi")),
    "top2mass"   : Hist(hist.axis.Regular(bins=20, start=50, stop=250, name='top2mass', label="top2mass")),
}

powheg_histo_dict = {
    "avg_top_pt" : Hist(hist.axis.Regular(bins=20, start=0, stop=400, name="avg_top_pt", label='avg_top_pt')),
    "mtt"        : Hist(hist.axis.Regular(bins=75, start=0, stop=1500, name="mtt", label='mtt')),
    "top1pt"     : Hist(hist.axis.Regular(bins=40, start=0, stop=800, name="top1pt", label='top1pt')),
    "top1eta"    : Hist(hist.axis.Regular(bins=40, start=-4, stop=4, name='top1eta', label='top1eta')),
    "top1phi"    : Hist(hist.axis.Regular(bins=40, start=-4, stop=4, name='top1phi', label="top1phi")),
    "top1mass"   : Hist(hist.axis.Regular(bins=20, start=50, stop=250, name='top1mass', label="top1mass")),
    "top2pt"     : Hist(hist.axis.Regular(bins=40, start=0, stop=800, name="top2pt", label='top2pt')),
    "top2eta"    : Hist(hist.axis.Regular(bins=40, start=-4, stop=4, name='top2eta', label='top2eta')),
    "top2phi"    : Hist(hist.axis.Regular(bins=40, start=-4, stop=4, name='top2phi', label="top2phi")),
    "top2mass"   : Hist(hist.axis.Regular(bins=20, start=50, stop=250, name='top2mass', label="top2mass")),
}
```

```python
smeft_variables_to_fill = {
    "avg_top_pt" : inputs_smeft['avg_top_pt'].to_numpy(),
    "mtt"        : inputs_smeft['mtt'].to_numpy(),
    "top1pt"     : inputs_smeft['top1pt'].to_numpy(),
    "top1eta"    : inputs_smeft['top1eta'].to_numpy(),
    "top1phi"    : inputs_smeft['top1phi'].to_numpy(),
    "top1mass"   : inputs_smeft['top1mass'].to_numpy(),
    "top2pt"     : inputs_smeft['top2pt'].to_numpy(),
    "top2eta"    : inputs_smeft['top2eta'].to_numpy(),
    "top2phi"    : inputs_smeft['top2phi'].to_numpy(),
    "top2mass"   : inputs_smeft['top2mass'].to_numpy(),
}

powheg_variables_to_fill = {
    "avg_top_pt" : inputs_powheg['avg_top_pt'].to_numpy(),
    "mtt"        : inputs_powheg['mtt'].to_numpy(),
    "top1pt"     : inputs_powheg['top1pt'].to_numpy(),
    "top1eta"    : inputs_powheg['top1eta'].to_numpy(),
    "top1phi"    : inputs_powheg['top1phi'].to_numpy(),
    "top1mass"   : inputs_powheg['top1mass'].to_numpy(),
    "top2pt"     : inputs_powheg['top2pt'].to_numpy(),
    "top2eta"    : inputs_powheg['top2eta'].to_numpy(),
    "top2phi"    : inputs_powheg['top2phi'].to_numpy(),
    "top2mass"   : inputs_powheg['top2mass'].to_numpy(),
}

for var_name, var_values in smeft_variables_to_fill.items():
    smeft_histo_dict[var_name].fill(var_values)
    
for var_name, var_values in powheg_variables_to_fill.items():
    powheg_histo_dict[var_name].fill(var_values)
```

```python
for name in smeft_histo_dict: 
    hep.style.use("CMS")
    fig, ax = plt.subplots()
    hep.histplot(smeft_histo_dict[name], ax=ax, yerr=False, label="smeft")
    hep.histplot(powheg_histo_dict[name], ax=ax, yerr=False, label="powheg")
    
    ax.legend()
    fig.show()
```
